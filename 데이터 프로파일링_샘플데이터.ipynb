## 로드된 파일 확인

```python
new_path = '/content/drive/MyDrive/Colab Notebooks/FacialClassification/'
```

```python
!ls
```

```python
with open(new_path+'temp.json', 'w', encoding='utf-8') as file:
    json.dump(temp, file, indent='\t')

    
for j in range(0,len(json_list),1):
    a=0
    miss=0
    path=os.path.join(new_path, sample_cate[j])+'/'
    print(sample_cate[j],'Annotation 탐색 시작')
    for i in json_list[j]:
        # print(i['filename'])
        # print(i['annot_A']['faceExp'],i['annot_B']['faceExp'],i['annot_C']['faceExp'], sample_cate[j])
        if not i['annot_A']['faceExp']==i['annot_B']['faceExp']==i['annot_C']['faceExp']==sample_cate[j]:
            temp.append(i)
            try:
                shutil.move(path+i['filename'], new_path+'temp/')
                print(i['filename']+'이 temp로 옮겨졌습니다.')

            except Exception as e:
                # print(i['filename'],'파일이 존재하지 않습니다')
                print(e)
                miss+=1
            # shutil.move(path+i['filename'], new_path+'temp/')
            a+=1
    for i in temp :
        json_list[j].remove(i)
    print(f'{sample_cate[j]} 카테고리 중 검수 필요한 파일의 수 : {a}')
    print(sample_cate[j],'카테고리 중 누락된 파일의 수 :',miss)

print(f'전체 검수가 필요한 샘플의 수 : {len(temp)}')
with open(f'{new_path}temp.json','w', encoding='utf-8') as file:
    json.dump(temp,file, indent='\t')
```

```python

for i in temp:
    new_cate=''
    if i.get('annot_A').get('faceExp')==i.get('annot_B').get('faceExp')==i.get('annot_C').get('faceExp') :
        new_cate = i.get('annot_A').get('faceExp')
        print(i.get('annot_A').get('faceExp'), i.get('annot_B').get('faceExp'), i.get('annot_C').get('faceExp'))
        print(new_cate)

        path='./FacialClassification/temp/'
        new_path='./FacialClassification/'+new_cate+'/'
        shutil.move(path+i.get('filename'), new_path+i.get('filename'))
        print(i['filename']+'파일이',new_cate,'폴더로 옮겨졌습니다')

        if new_cate=='기쁨':
            json_list[0].append(i)
        elif new_cate=='당황':
            json_list[1].append(i)
        elif new_cate=='분노':
            json_list[2].append(i)
        elif new_cate=='불안':
            json_list[3].append(i)
        elif new_cate=='상처':
            json_list[4].append(i)
        elif new_cate=='슬픔':
            json_list[5].append(i)
        elif new_cate=='중립':
            json_list[6].append(i)
        else:
            print('오류가 있습니다')
            print(i.get('filename'))

    else : 
        print(i['filename'],' 파일을 보류합니다')
    
```
