```python
def unzip(source_file, dest_path):
    with zipfile.ZipFile(source_file, 'r') as zf:
        zipInfo = zf.infolist()
        c=0
        w=len(zipInfo)
        for member in zipInfo:
            try:
                c+=1
                print('\r', int(c/w*100),'%', '  ', c,'/',w, member.filename.encode('cp437').decode('euc-kr', 'ignore'), end='')

                # print(member.filename.encode('cp437').decode('euc-kr', 'ignore'))
                member.filename = member.filename.encode('cp437').decode('euc-kr', 'ignore')
                zf.extract(member, dest_path)
            except:
                print(source_file)
                raise Exception('what?!')
```

```python
cl=['기쁨','당황','불안','상처']
```

```python
for cate in cl:
    for i in range(4):
        print(f'{cate} 압축 푸는중')
        unzip(f'./[원천]EMOIMG_{cate}_TRAIN_0{i+1}.zip', f'./FacialClassification/{cate}')
    print(f'{cate} 압축해제 완료')
```
